### Q5 - Engine HP

```python
avg_hp_before = df["Engine HP"].mean()
avg_hp_before
```

```python
avg_hp_after = df["Engine HP"].fillna(avg_hp_before).mean()
avg_hp_after
```

```python
if round(avg_hp_after) == round(avg_hp_before):
    print("There was no change between 'before' and 'after'")
else:
    print("There was a change between in Engine HP")
```

### Q6 - matrix operations on RollsRoyce

```python
#all Rolls Royces
df[df["Make"]== "ROLLS-ROYCE"]
```

```python
# only selected columns
rrdf = df[df["Make"]== "ROLLS-ROYCE"][["Engine HP", "Engine Cylinders", "highway MPG"]]
rrdf
```

```python
rrdf_dedup =rrdf.drop_duplicates()
rrdf_dedup
```

```python
#getting np array
X=rrdf_dedup.values
X
```

```python
X_T = X.T
XTX = X_T.dot(X)
XTX
```

```python
np.matmul(X_T,X)
```

```python
XTX_inv=np.linalg.inv(XTX)
XTX_inv
```

```python
XTX_inv.sum()
```
